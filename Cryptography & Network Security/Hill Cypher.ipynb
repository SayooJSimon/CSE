```python
import numpy as np

def get_key_matrix(key):
   key_matrix = np.zeros((3, 3), dtype=int)
   k = 0
   for i in range(3):
       for j in range(3):
           key_matrix[i][j] = ord(key[k]) % 65
           k += 1
   return key_matrix

def encrypt(message, key_matrix):
   message_vector = np.array([[ord(c) % 65] for c in message])
   cipher_matrix = np.dot(key_matrix, message_vector) % 26
   return ''.join(chr(int(num) + 65) for num in cipher_matrix)

def hill_cipher(message, key):
   key_matrix = get_key_matrix(key)
   return encrypt(message, key_matrix)
```

```python
message = input("Enter the plain text: ")
key = "GYBNQKURP"
ciphertext = hill_cipher(message, key)
print("Ciphertext:", ciphertext)
```
