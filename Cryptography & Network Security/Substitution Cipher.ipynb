```python
import random
import string

def generate_cipher_map():
    letters = list(string.ascii_uppercase)
    shuffled = letters[:]
    random.shuffle(shuffled)  # Shuffle letters to create mapping
    cipher_map = dict(zip(letters, shuffled))  # Map original letters to shuffled letters
    return cipher_map

cipher_map = generate_cipher_map()
print(cipher_map)

def encode(text, cipher):
   return "".join(cipher.get(ch.upper(), ch) for ch in text)

def decode(text, cipher):
   reverse_cipher = {v: k for k, v in cipher.items()}
   return "".join(reverse_cipher.get(ch.upper(), ch) for ch in text)
# Example cipher mapping
```

```python
plaintext = input("Enter the plaintext: ")
encrypted = encode(plaintext, cipher_map)
decrypted = decode(encrypted, cipher_map)
print("Encrypted:", encrypted)
print("Decrypted:", decrypted)
```
