```python
import numpy as np

def mod_inverse(a, m):
    return pow(a, -1, m)

def matrix_mod_inverse(matrix, modulus):
    det = int(np.round(np.linalg.det(matrix)))
    det_mod = det % modulus
    det_inv = mod_inverse(det_mod, modulus)
    adjugate = np.round(det * np.linalg.inv(matrix)).astype(int)
    return (det_inv * adjugate) % modulus

def text_to_numbers(text):
    return [ord(char) - ord('A') for char in text]

def numbers_to_text(numbers):
    return ''.join(chr(int(num) % 26 + ord('A')) for num in numbers)

def hill_encrypt(plaintext, key_matrix):
    plaintext = plaintext.replace(" ", "").upper()
    n = key_matrix.shape[0]

    while len(plaintext) % n != 0:
        plaintext += 'X'

    plaintext_nums = text_to_numbers(plaintext)
    ciphertext = ""

    for i in range(0, len(plaintext_nums), n):
        block = np.array(plaintext_nums[i:i+n])
        encrypted_block = np.dot(key_matrix, block) % 26
        ciphertext += numbers_to_text(encrypted_block)

    return ciphertext

def hill_decrypt(ciphertext, key_matrix):
    n = key_matrix.shape[0]
    key_inverse = matrix_mod_inverse(key_matrix, 26)

    ciphertext_nums = text_to_numbers(ciphertext)
    plaintext = ""

    for i in range(0, len(ciphertext_nums), n):
        block = np.array(ciphertext_nums[i:i+n])
        decrypted_block = np.dot(key_inverse, block) % 26
        plaintext += numbers_to_text(decrypted_block)

    return plaintext

if __name__ == "__main__":
    key = np.array([[3, 3],
                    [2, 5]])

    message = "WORLD"

    encrypted = hill_encrypt(message, key)
    print("Encrypted:", encrypted)

    decrypted = hill_decrypt(encrypted, key)
    print("Decrypted:", decrypted)
```
